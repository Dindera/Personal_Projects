```python
# Inspect some categorical columns, AGE,MARRIED, VEHICLE_OWNERSHIP,CHILDREN,VEHICLE_YEAR, OUTCOME, POSTAL_CODE, VEHICLE_TYPE
cat_cols = [df['AGE'].value_counts(), 
            df['MARRIED'].value_counts(),
            df['VEHICLE_OWNERSHIP'].value_counts(),
            df['CHILDREN'].value_counts(), 
            df['VEHICLE_YEAR'].value_counts(),
            df['POSTAL_CODE'].value_counts(),
            df['VEHICLE_TYPE'].value_counts(),
            df['OUTCOME'].value_counts()
           ]
cat_cols
```

```python
# Change datatype and category of some tables VEHICLE_OWNERSHIP, MARRIED, CHILDREN, OUTCOME
df['VEHICLE_OWNERSHIP'] = df['VEHICLE_OWNERSHIP'].replace([1.0, 0.0], ['Vehicle Owner', 'Not Vehicle Owner'])
df['MARRIED'] = df['MARRIED'].replace([1.0, 0.0], ['Married', 'Not Married'])
df['CHILDREN'] = df['CHILDREN'].replace([1.0, 0.0], ['Yes', 'No'])
df['OUTCOME'] = df['OUTCOME'].replace([1.0, 0.0], ['Yes', 'No'])
```

```python
df.head(5)
```

```python
# Change datatype of annual mileage to int and reduce decimal place of credit_score to 2
df['ANNUAL_MILEAGE'] = df['ANNUAL_MILEAGE'].astype(int)
df['CREDIT_SCORE'] = df['CREDIT_SCORE'].round(decimals=2)
```

```python
# Remove preceding y in DRIVING_EXPERIENCE
df['DRIVING_EXPERIENCE'] = df['DRIVING_EXPERIENCE'].str.split('y').str.get(0)
```

```python
df.head(3)
```

```python
# Convert and rename columns to lowercase except first letter
df_columns = [df.columns[i].title() for i in range(len(df.columns))]
df.columns.values[0:19] = df_columns
```

```python
df.head(3)
```

```python
#Rename Age to Age_Group
df = df.rename(columns={'AGE': 'Age_Group'})
df.head(2)
```

```python
# Save clean data for future use
df.to_csv('Data/Car_Insurance_Claim_Output.csv')
```
